```python
import pandas as pd
import math
```

```python
#Labels for dataset
labels = ['Feature ' + str(i) for i in range(23)]

#Reading training dataset from excel sheet
training_data = pd.read_excel('parktraining.xlsx', header=None, names=labels)

#Seperating Class Disease(last column) from data set, to apply feature scaling on remaining data set
Disease = pd.DataFrame({'Disease': training_data.iloc[:,-1]})
training_data = training_data.drop(columns='Feature 22')

#Feature Scaling
min_ = training_data.min()
max_ = training_data.max()

for i in range(len(training_data)):
    for j in range(len(training_data.columns)):
        training_data.iloc[i, j] = (training_data.iloc[i, j] - min_[j]) / (max_[j] - min_[j])
        
#Rejoin class Disease to the dataset
training_data = training_data.join(Disease)
```

```python
#Seperating classes wrt disease

disease_pd = training_data[training_data['Disease'] == 1]
healthy = training_data[training_data['Disease'] == 0] 
```

```python
#Calculating Prior probabilities of each class

prob_pd = len(disease_pd)/len(training_data)
prob_healthy = len(healthy)/len(training_data)
```

```python
# Calculating mean of each class for each feature

mean_pd = disease_pd.mean()
mean_healthy = healthy.mean()
```

```python
# Calculating variance of each class for each feature

variance_pd = disease_pd.var()
variance_healthy = healthy.var()
```

```python
#Calculating standard deviation

std_pd = disease_pd.std()
std_healthy = healthy.std()
```

```python
#Labels for dataset
labels_ = ['Feature ' + str(i) for i in range(23)]

#Reading training dataset from excel sheet
testing_data = pd.read_excel('parktesting.xlsx', header=None, names=labels_)

#Seperating Class Disease(last column) from data set, to apply feature scaling on remaining data set
actual_label = pd.DataFrame({'Actual Label': testing_data.iloc[:,-1]})
testing_data = testing_data.drop(columns='Feature 22')

#Feature Scaling
min_ = testing_data.min()
max_ = testing_data.max()

for i in range(len(testing_data)):
    for j in range(len(testing_data.columns)):
        testing_data.iloc[i, j] = (testing_data.iloc[i, j] - min_[j]) / (max_[j] - min_[j])
```

```python
predicted_label = []
pi = 3.14159265

for i in range(len(testing_data)):
    
    label_wrt_class_healthy = 1       #healthy
    label_wrt_class_pd = 1            #unhealthy
    
    denominator = 0
    
    for j in range(len(testing_data.columns)):
        
        label_wrt_class_healthy *= (math.exp(-((testing_data.iloc[i,j] - mean_healthy[j])**2 / (2*variance_healthy[j]))))
        label_wrt_class_healthy /= math.sqrt(2*pi)*std_healthy[j]      
        
        label_wrt_class_pd *= (math.exp(-((testing_data.iloc[i,j] - mean_pd[j])**2 / (2*variance_pd[j]))))
        label_wrt_class_pd /= math.sqrt(2*pi)*std_pd[j]       
     
    label_wrt_class_healthy *= prob_healthy
    label_wrt_class_pd *= prob_pd
    
    denominator = label_wrt_class_healthy + label_wrt_class_pd
    
    label_wrt_class_healthy /= denominator
    label_wrt_class_pd /= denominator
    
    if label_wrt_class_healthy > label_wrt_class_pd:
        predicted_label.append(0)
    else:
        predicted_label.append(1)
        
```

```python
predicted_labels = pd.DataFrame({'Predicted Label' : predicted_label})
label = actual_label.join(predicted_labels)
label
```

```python
correct = 0
incorrect = 0

for i in range(len(label)):
    if label.iloc[i,0] == label.iloc[i, 1]:
        correct += 1
    else:
        incorrect += 1
        
print('Correct :',correct)
print('Incorrect :',incorrect)
print('Accuracy :',correct/len(label))
```
